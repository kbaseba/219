```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_train = 200
x = np.random.uniform(low = 0, high = 1, size = (num_train, ))
y = 2 * x + 3 + np.random.normal(loc = 0, scale = 0.1, size = (num_train, ))
f = plt.figure()
ax = f.gca()
ax.plot(x, y, '.')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```

Equation: $y = 2x + 3 + \epsilon$


### 1. Represent the data

$y = X \theta + \epsilon$

$y$ is the vector of observed values: $y = [y_1, y_2, \dots, y_n]^T \in \mathbf{R}^{N}$,

$X$ is the input matrix: $X = [\mathbf{x}, \mathbf{1}] \in \mathbf{R}^{N \times 2}$

### 2. Objective function

$L(\theta) = \|y - X \theta\|^2$

### 3. Partial derivatives and simplify

$\frac{\partial L}{\partial \theta} = -2 X^T (y - X \theta) = 0$

$X^T X \theta = X^T y$

$\theta = (X^T X)^{-1} X^T y$



$A^T$

```python
A = np.array([[1,2],
              [3,4],
              [5,6]])
print(A.T)
```

$A^TA$

```python
print(A.T.dot(A))
```

$(A^TA)^{-1}$

```python
print(np.linalg.inv(A.T.dot(A)))
```

```python
xhat = np.stack((x, np.ones_like(x)), axis =1)

theta = np.linalg.inv(xhat.T.dot(xhat)).dot(xhat.T.dot(y)) # solution

print(theta)
```

```python
f = plt.figure()
ax = f.gca()
ax.plot(x, y, '.')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

# Plot the regression line
xs = np.linspace(min(x), max(x),50)
xs = np.vstack((xs, np.ones_like(xs)))
plt.plot(xs[0,:], theta.dot(xs))
```
